Now, let's try a custom dataset of helicopters that I found online, compared to a different dataset of bananas (I chose very different objects to make training times faster). If you are trying to run this, you may have to redownload the data; they're both from https://images.cv, but the actual datasets shouldn't matter. You will also need to remake the custom.yaml file, I used the following tutorial: https://docs.ultralytics.com/yolov5/tutorials/train_custom_data/#11-create-datasetyaml.

```python
# to make label values

import os
rootdir = 'datasets/helicopters/images'

for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        print(os.path.join(subdir, file))
        with open(os.path.join(subdir, file).replace("images", "labels").replace("jpg", "txt"), "w") as f:
          f.write("1 0 0 1 1") # don't care about bounding box, so we make it fill screen
```

```python
# now, the fun part

from ultralytics import YOLO

# Load a pretrained YOLO11n model
model = YOLO("yolo11n.pt")

# Train the model on COCO8
results = model.train(data="custom.yaml", epochs=100, imgsz=64, resume=True)
```

In the end, we reached a P value of around 50%, which doesn't seem great, until one realizes that the P value corresponds to the box, which we don't care about. We actually care about the R value, which corresponds to classification proportion. For this, we did very well, with an A- (~90%)! The entire training (100 epochs) took around an hour on Google Colab.
